### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
df_filtered = df[['JobSatPoints_6','JobSatPoints_7','Age']].dropna() 

df_grouped = df_filtered.groupby('Age')[['JobSatPoints_6','JobSatPoints_7']].sum()

fig, ax = plt.subplots(figsize = (16,12)) 

age_groups = df_grouped.index
bar_width = 0.6 

ax.bar(age_groups, df_grouped['JobSatPoints_6'], label = "Job Satisfaction Points (6)", color = "blue")
ax.bar(age_groups, df_grouped['JobSatPoints_7'], label = "Job Satisfaction Points (7)", color = "green") 

ax.set_xlabel("Age Groups") 
ax.set_ylabel("Job Satisfaction Points") 
ax.set_title("Job Satisfaction Composition Across Different Age Groups") 

plt.show()
```

##### Stacked Chart of `JobSatPoints_6` and `JobSatPoints_7` for Employment Status


Create a stacked chart to compare job satisfaction (`JobSatPoints_6` and `JobSatPoints_7`) across different employment statuses. This will show how satisfaction varies by employment type.


```python
df_filtered_1 = df[['JobSatPoints_6','JobSatPoints_7','Employment']].dropna() 

df_grouped_1 = df_filtered_1.groupby('Employment')[['JobSatPoints_6','JobSatPoints_7']].sum()

fig, ax = plt.subplots(figsize = (16,12)) 

emp_groups = df_grouped_1.index
bar_width = 0.6 

ax.bar(emp_groups, df_grouped_1['JobSatPoints_6'], label = "Job Satisfaction Points (6)", color = "blue")
ax.bar(emp_groups, df_grouped_1['JobSatPoints_7'], label = "Job Satisfaction Points (7)", color = "green") 

ax.set_xlabel("Employment Types") 
ax.set_ylabel("Job Satisfaction Points") 
ax.set_title("Job Satisfaction Composition Across Different Employment Types") 

plt.xticks(rotation = 90)
plt.show()
```

### Task 2: Stacked Chart for Compensation and Job Satisfaction by Age Group


##### This stacked chart visualizes the composition of compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) specifically for respondents aged 30-35.


```python
age_mapping = {
    "Under 18 years old": 15,
    "18-24 years old": 21,
    "25-34 years old": 29.5,
    "35-44 years old": 39.5,
    "45-54 years old": 49.5,
    "55-64 years old": 59.5,
    "65 years or older": 70,
    "Prefer not to say": None
}

df['Age_numeric'] = df['Age'].map(age_mapping) 

df_filtered_2 = df[df['Age'].isin(['25-34 years old','35-44 years old'])] 

df_filtered_2 = df_filtered_2.dropna(subset = ['ConvertedCompYearly','JobSatPoints_6','Age']) 

df_grouped_2 = df_filtered_2.groupby('Age_numeric')[['ConvertedCompYearly','JobSatPoints_6']].sum() 

fig, ax = plt.subplots(figsize = (16, 12)) 

age_groups = df_grouped_2.index 
bar_width = 0.6 

ax.bar(age_groups, df_grouped_2['ConvertedCompYearly'], label = 'ConvertedCompYearly', color = 'blue') 
ax.bar(age_groups, df_grouped_2['JobSatPoints_6'], label = 'Job Satisfaction Points (6)', bottom = df_grouped_2['ConvertedCompYearly'], color = 'green') 

ax.set_xlabel("Age Groups") 
ax.set_ylabel("Sum of ConvertedCompYearly and Job Satisfaction Points (6)") 
ax.set_title("Composition of ConvertedCompYearly and Job Satisfaction Points (6)") 

plt.show()
```

##### Stacked Chart of Median Compensation and Job Satisfaction Across Age Group


Compare the median compensation and job satisfaction metrics across different age groups. This helps visualize how compensation and satisfaction levels differ by age.


```python
df_filtered_3 = df[['JobSat','ConvertedCompYearly','Age']].dropna() 

df_grouped_median = df_filtered_3.groupby('Age')[['ConvertedCompYearly','JobSat']].median()

fig, ax = plt.subplots(figsize = (16,12)) 

age_groups = df_grouped_median.index
bar_width = 0.6 

ax.bar(age_groups, df_grouped_median['ConvertedCompYearly'], label = "ConvertedCompYearly", color = "blue")
ax.bar(age_groups, df_grouped_median['JobSat'], label = "Job Satisfaction", color = "green") 

ax.set_xlabel("Age Groups") 
ax.set_ylabel("Job Satisfaction and ConvertedCompYearly") 
ax.set_title("Job Satisfaction and ConvertedCompYearly Composition Across Different Age Groups") 

plt.show()
```

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
df_filtered_4 = df[['DatabaseWantToWorkWith','Age']].dropna() 

df_filtered_4 = df_filtered_4.assign(DatabaseWantToWorkWith = df_filtered_4['DatabaseWantToWorkWith'].str.split(';')).explode('DatabaseWantToWorkWith')

df_grouped_4 = df_filtered_4.groupby(['Age','DatabaseWantToWorkWith']).size().unstack(fill_value = 0)

fig, ax = plt.subplots(figsize = (16,12)) 

df_grouped_4.plot(kind = 'bar', stacked = True, ax=ax, colormap = "tab20", width = 0.6)

ax.set_xlabel("Age Groups") 
ax.set_ylabel("Count of Preferred Database") 
ax.set_title("Count of Preferred Database Composition Across Different Age Groups") 

plt.show()
```

##### 2. Stacked Chart of Employment Type by Job Satisfaction


Analyze the distribution of employment types within each job satisfaction level using a stacked chart. This will provide insights into how employment types are distributed across various satisfaction ratings.


```python
df_filtered_5 = df[['Employment','JobSat']].dropna() 

df_grouped_5 = df_filtered_5.groupby(['JobSat','Employment']).size().unstack(fill_value = 0)

fig, ax = plt.subplots(figsize = (16,12)) 

df_grouped_5.plot(kind = 'bar', stacked = True, ax=ax, colormap = "tab20", width = 0.6)

ax.set_xlabel("Job Satisfaction Level") 
ax.set_ylabel("Count of Employment Type") 
ax.set_title("Count of Employment Type Composition Across Different Job Satisfaction Levels") 
ax.legend(title = "Employment Type", bbox_to_anchor = (1.05, 1), loc = 'best')
plt.show()
```

### Task 4: Exploring Technology Preferences Using Stacked Charts


##### 1. Stacked Chart for Preferred Programming Languages by Age Group


Analyze how programming language preferences (`LanguageAdmired`) vary across age groups.


```python
df_filtered_6 = df[['LanguageAdmired','Age']].dropna() 

df_filtered_6 = df_filtered_6.assign(LanguageAdmired = df_filtered_6['LanguageAdmired'].str.split(';')).explode('LanguageAdmired')

df_grouped_6 = df_filtered_6.groupby(['Age','LanguageAdmired']).size().unstack(fill_value = 0)

fig, ax = plt.subplots(figsize = (16,12)) 

df_grouped_6.plot(kind = 'bar', stacked = True, ax=ax, colormap = "tab20", width = 0.6)

ax.set_xlabel("Age Group") 
ax.set_ylabel("Count of Language Admired") 
ax.set_title("Count of Language Admired Across Different Age Groups") 
ax.legend(title = "Langugage", bbox_to_anchor = (1.05, 1), loc = 'best')
plt.show()
```

##### 2. Stacked Chart for Technology Adoption by Employment Type


Explore how admired platforms (`PlatformAdmired`) differ across employment types (e.g., full-time, freelance)


```python
df_filtered_7 = df[['PlatformAdmired','Employment']].dropna() 

df_filtered_7 = df_filtered_7.assign(LanguageAdmired = df_filtered_6['LanguageAdmired'].str.split(';')).explode('LanguageAdmired')

df_grouped_6 = df_filtered_6.groupby(['Age','LanguageAdmired']).size().unstack(fill_value = 0)

fig, ax = plt.subplots(figsize = (16,12)) 

df_grouped_6.plot(kind = 'bar', stacked = True, ax=ax, colormap = "tab20", width = 0.6)

ax.set_xlabel("Age Group") 
ax.set_ylabel("Count of Language Admired") 
ax.set_title("Count of Language Admired Across Different Age Groups") 
ax.legend(title = "Langugage", bbox_to_anchor = (1.05, 1), loc = 'best')
plt.show()
```
