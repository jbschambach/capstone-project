### Task 1: Visualizing Data Distributions


##### 1. Histogram of `ConvertedCompYearly`


Visualize the distribution of yearly compensation (`ConvertedCompYearly`) using a histogram.



```python
plt.figure(figsize = (12,10)) 
plt.hist(df['ConvertedCompYearly'].dropna(), bins = 30, edgecolor = 'black') 

plt.xlabel("ConvertedCompYearly") 
plt.ylabel("Distribution") 
plt.title("Distribution of ConvertedCompYearly") 

plt.show()
```

##### 2. Box Plot of `Age`


Since `Age` is categorical in the dataset, convert it to numerical values for a box plot.



```python
age_mapping = { 
    "Under 18 years old": 15,
    "18-24 years old": 21,
    "25-34 years old": 29.5,
    "35-44 years old": 39.5,
    "45-54 years old": 49.5,
    "55-64 years old": 59.5,
    "65 years or older": 70,
    "Prefer not to say": None
}

df['Age_numeric'] = df['Age'].map(age_mapping) 

plt.figure(figsize = (12,10)) 
plt.boxplot(df['Age_numeric'].dropna(), vert = True) 

plt.ylabel("Age") 
plt.title("Box Plot of Age") 

plt.show()
```

### Task 2: Visualizing Relationships in Data


##### 1. Scatter Plot of `Age_numeric` and `ConvertedCompYearly`


Explore the relationship between age and compensation.



```python
plt.figure(figsize = (12,10)) 
plt.scatter(df['Age_numeric'], df['ConvertedCompYearly'], alpha = 0.5) 

plt.xlabel("Age") 
plt.ylabel("ConvertedCompYearly")
plt.title("Scatter Plot of Age and Converted Compensation - Yearly") 

plt.show()
```

##### 2. Bubble Plot of `ConvertedCompYearly` and `JobSatPoints_6` with `Age_numeric` as Bubble Size


Explore how compensation and job satisfaction are related, with age as the bubble size.


```python
plt.figure(figsize = (12,10)) 
plt.scatter(
    df['JobSatPoints_6'], 
    df['ConvertedCompYearly'],
    s = df['Age_numeric']*5,
    alpha = 0.5) 

plt.xlabel("Job Satisfaction Points (6)") 
plt.ylabel("ConvertedCompYearly") 
plt.title("ConvertedCompYearly Across Job Satisfaction Points (6) (Bubble Size = Age)") 

plt.show()
```

### Task 3: Visualizing Composition of Data with Bar Charts


##### 1. Horizontal Bar Chart of `MainBranch` Distribution


Visualize the distribution of respondents’ primary roles to understand their professional focus.



```python
df['MainBranch'].value_counts().plot(kind = "barh", figsize = (12,10), color = "coral", edgecolor = "black") 

plt.xlabel("Count") 
plt.ylabel("Main Branches") 
plt.title("Count of Respondents in Main Branches") 

plt.show()
```

##### 2. Vertical Bar Chart of Top 5 Programming Languages Respondents Want to Work With


Identify the most desired programming languages based on `LanguageWantToWorkWith`.



```python
lang_list = df['LanguageWantToWorkWith'].dropna().str.split(';').explode() 

top_5 = lang_list.value_counts().head(5) 

top_5.plot(kind = "bar", figsize = (12,10), color = 'skyblue', edgecolor = 'black')

plt.xlabel("Programming Language") 
plt.ylabel("Count") 
plt.title("Top 5 Programming Languages Respondents Want To Work With") 

plt.show()
```

##### 3. Stacked Bar Chart of Median `JobSatPoints_6` and `JobSatPoints_7` by Age Group


Compare job satisfaction metrics across different age groups with a stacked bar chart.


```python
df_median_points = df.groupby('Age_numeric')[['JobSatPoints_6','JobSatPoints_7']].median() 

ax = df_median_points.plot(kind = 'bar', stacked = True, figsize = (12,10), color = ['blue','yellow']) 

plt.xlabel("Age") 
plt.ylabel("Median Job Satisfaction Points") 
plt.title("Job Satisfaction Points (6 & 7) Across Age Groups") 
plt.show()
```

##### 4. Bar Chart of Database Popularity (`DatabaseHaveWorkedWith`)


Identify the most commonly used databases among respondents by visualizing `DatabaseHaveWorkedWith`.



```python
databases = df['DatabaseHaveWorkedWith'].dropna().str.split(';').explode() 

database_counts = databases.value_counts() 

top_10 = database_counts.head(10) 

plt.figure(figsize = (12,10)) 
top_10.plot(kind = "bar", color = "purple") 

plt.xlabel("Database") 
plt.ylabel("Count of Respondents") 
plt.title("Top 10 Popular Databases Respondents Have Worked With") 

plt.show()
```

### Task 4: Visualizing Comparison of Data with Bar Charts


##### 1. Grouped Bar Chart of Median `ConvertedCompYearly` for Different Age Groups


Compare median compensation across multiple age groups with a grouped bar chart.



```python
median_comp = df.groupby('Age_numeric')['ConvertedCompYearly'].median() 

plt.figure(figsize = (12,10)) 
median_comp.plot(kind = "bar", color = "green") 

plt.xlabel("Age") 
plt.ylabel("Median ConvertedCompYearly") 
plt.title("Median ConvertedCompYearly by Age") 

plt.show()
```

##### 2. Bar Chart of Respondent Count by Country


Show the distribution of respondents by country to see which regions are most represented.



```python
country_count = df['Country'].value_counts() 

country_count.plot(kind = 'bar', figsize = (50,30), color = 'orange') 

plt.xlabel("Country")
plt.xticks(rotation = 90)
plt.ylabel("Number of Respondents") 
plt.title("Number of Respondents per Country") 

plt.show()
```
